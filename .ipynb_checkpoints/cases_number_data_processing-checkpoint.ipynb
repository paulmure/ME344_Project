```python
import pandas as pd
```

```python
raw_data = pd.read_csv('us-counties.csv')
raw_data = raw_data[pd.notna(raw_data['fips'])]
# convert dates to indeces
dates = raw_data['date'].tolist()
seen = set()
seen_add = seen.add
unique_dates = [x for x in dates if not(x in seen or seen_add(x))]
date_to_index = {date: idx for idx, date in enumerate(unique_dates)}

"""fips = raw_data['fips'].tolist()
seen = set()
seen_add = seen.add
unique_fips = [x for x in fips if not(x in seen or seen_add(x))]
fips_to_index = {fips: idx for idx, fips in enumerate(unique_fips)}"""
```

```python
raw_data['date_index'] = raw_data.apply(lambda row: date_to_index[row['date']], axis=1)
raw_data['fips_index'] = raw_data.apply(lambda row: int(row['fips']), axis=1)
print(raw_data['fips_index'])
```

```python
cols = ['date_index', 'fips_index', 'cases', 'deaths']
data = raw_data[cols]
data
```

```python
data.to_csv('processed_cases.csv', index=False, header=False)
```

```python
df = pd.read_csv('Average_Household_Size_and_Population_Density_-_County.csv')
df
```

```python
house_size = {}
pop_dens = {}
total_pop = {}
for i in range(len(df)):
    house_size[df.loc[i, 'GEOID']] = df.loc[i, 'B25010_001E']
    pop_dens[df.loc[i, 'GEOID']] = df.loc[i, 'B01001_calc_PopDensity']
    total_pop[df.loc[i, 'GEOID']] = df.loc[i, 'B01001_001E']
```

```python
unique_fips = list(house_size.keys())
```

```python
def see_if_fips_exists(row):
    exist = row['fips_index'] in unique_fips
    return int(exist)
data['exists'] = data.apply(see_if_fips_exists, axis=1)
```

```python
data = data[data['exists'] == 1]
data['house_size'] = data.apply(lambda row: house_size[row['fips_index']], axis=1)
data['pop_dens'] = data.apply(lambda row: pop_dens[row['fips_index']], axis=1)

data['total_pop'] = data.apply(lambda row: total_pop[row['fips_index']], axis=1)
data
```

```python
data.columns
```

```python
cols = ['date_index', 'cases', 'house_size', 'pop_dens', 'total_pop', 'deaths']
data = data[cols]
data.to_csv('processed_cases.csv', index=False, header=False)
```
