```python
from tensorflow import keras
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv('us-counties.csv')

# convert dates to indeces
dates = raw_data['date'].tolist()
seen = set()
seen_add = seen.add
unique_dates = [x for x in dates if not(x in seen or seen_add(x))]
date_to_index = {date: idx for idx, date in enumerate(unique_dates)}

date_to_index
```

```python
raw_data['date_index'] = raw_data.apply(lambda row: date_to_index[row['date']], axis=1)
```

```python
cols = ['date_index', 'fips', 'cases', 'deaths']
data = raw_data[cols]
data
```

```python
data.to_csv()
```
